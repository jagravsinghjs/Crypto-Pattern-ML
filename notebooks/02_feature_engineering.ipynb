```python
import pandas as pd

df = pd.read_pickle("../data/processed/cleaned.pkl")
df.head()
```

```python
df['ma_5'] = df['Close'].rolling(5).mean()
df['ma_10'] = df['Close'].rolling(10).mean()
```

```python
df['momentum_5'] = df['Close'] - df['Close'].shift(5)
```

```python
df['volatility_10'] = df['return'].rolling(10).std()
```

```python
df['volume_norm'] = (df['Volume'] - df['Volume'].mean()) / df['Volume'].std()
```

```python
for i in range(1, 6):
    df[f'return_lag_{i}'] = df['return'].shift(i)
```

```python
features = [
    'return',
    'ma_5', 'ma_10',
    'momentum_5',
    'volatility_10',
    'volume_norm',
    'return_lag_1', 'return_lag_2', 'return_lag_3',
    'return_lag_4', 'return_lag_5'
]
```

```python
df = df.dropna()
df.to_pickle("../data/processed/features.pkl")
print("features.pkl saved!")
```
