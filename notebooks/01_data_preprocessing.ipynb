```python
import pandas as pd

df = pd.read_csv("../data/raw/Beat_The_Market_Dataset.csv")
df.head()
```

```python
import numpy as np

df['return'] = np.log(df['Close'] / df['Close'].shift(1))
df['return']
```

```python
df['future_return'] = df['return'].shift(-1)
df['y'] = (df['future_return'] > 0).astype(int)

df['future_return']
```

```python
df['y'].value_counts()
```
